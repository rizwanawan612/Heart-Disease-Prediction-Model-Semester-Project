```python
#loading required packages
import pandas as pd
import matplotlib.pyplot as plt
#plt.style.use('classic')

%matplotlib inline
```

```python
# loading dataset
df=pd.read_csv('D:/MS DS/TT for DS Theory/data set of cardio disease/heart_statlog_cleveland_hungary_final.csv')
#df=data.copy()
df.head()
```

```python
df.columns
```

```python
#changing columns names
df.columns = ['age', 'sex', 'chestpaintype', 'restingbps', 'cholesterol',
       'fasting blood sugar', 'resting ecg', 'max heart rate',
       'exercise angina', 'oldpeak', 'ST slope', 'target']
df.head()
```

```python
# Aggregating population on year
df_popworld=df.groupby(['age'])['cholesterol'].sum().reset_index()
df_popworld.head()
```

```python
# Make a line plot: year on the x-axis, pop on the y-axis
age=df_popworld['age']
cholesterol=df_popworld['cholesterol']
plt.plot(age,cholesterol,"go--",color='cyan')
# Display the plot with plt.show() IT HELPS IN FLUSHING THE MEMORY
plt.show()
```

```python
# Make a line plot, v on the x-axis, cholesterol on the y-axis  

plt.plot(df['age'],df['cholesterol'], "y")
plt.show()
```

```python
# Aggregating age on restingbps
df_popworl=df.groupby(['age'])['restingbps'].sum().reset_index()
df_popworl.head()
```

```python
# Make a line plot: age on the x-axis, restingbps on the y-axis
age=df_popworl['age']
restingbps=df_popworl['restingbps']
plt.plot(age,restingbps,"go--",color='cyan')
# Display the plot with plt.show() IT HELPS IN FLUSHING THE MEMORY
plt.show()
```

```python
# Make a line plot, age on the x-axis, restingbps on the y-axis  

plt.plot(df['age'],df['restingbps'], "y")
plt.show()
```

```python
# Make a scatter plot, age on the x-axis, cholesterol on the y-axis
plt.scatter(df['age'],df['cholesterol'])
plt.show()
```

```python
# Make a scatter plot, age on the x-axis, restingbps on the y-axis
plt.scatter(df['age'],df['restingbps'])
plt.show()
```

```python
# Put the x-axis on a logarithmic scale
plt.scatter(df['age'],df['restingbps'], color='g', alpha=1)
plt.xscale('log')
plt.show()
```
